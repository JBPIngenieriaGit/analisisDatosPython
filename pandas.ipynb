```python
import pandas as pd
```

```python
# Listas y Series en pandas, son estructuras unidimensionales, solo una columna
#crear una lista
numeros = [3, 4, 5, 6, 7]
# creamos la serie con la lista numeros
serie = pd.Series(numeros)
# se imprime
serie, type(serie)
```

```python
# Diccionario de datos, contiene claves y valores, como un json
data = {
    "Nombre":["julio", "camilo", "ramiro", "camila"],
    "Edad":[22, 35, 67, 12],
    "Color":["rojo", "verde", "amarillo", "negro"]
}
data, type(data)
```

```python
# Crear un data Frame, estructura bidimensional, como una tabla, con multiples columnas 
# para el ejemplo se construye el dataFrame a partir del diccionario creado anteriormente, cada clave sera una \
# columna y su valor sera un elemento de esa columna

df = pd.DataFrame(data=data) 
df


  
```
