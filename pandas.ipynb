TRATAMIENTO DE DATOS CON PANDS y NUMPY VIDEO 6

```python
# Importar numpy
import numpy as np
```

```python
# Crear un df con valores null en campos de texto y campos numericos,
# estos valores deben ser tratados antes de aplicar cualquier tecnica de Machine Learning (missing value o imputación de datos)

data = {
    "Nombre":["julio", "camilo", "ramiro", "camila"],
    "Edad":[22, np.nan, 67, 12],
    "Color":["rojo", "verde", None, "negro"]
}
df = pd.DataFrame(data)
df
```

Los missing values (valores faltantes o NA's) son datos ausentes, nulos o no registrados en un conjunto de información (dataset), comunes en encuestas o errores de registro. La imputación de datos es el proceso de reemplazar estos huecos por valores simulados o estimados (como la media o mediana) para no perder información valiosa ni sesgar el análisis estadístico. 

Aspectos clave:
¿Por qué ocurren? Pueden ser errores de transcripción, encuestados que no responden, o información no aplicable.
Tipos de faltantes:
MCAR (Completamente al azar): La falta de datos no tiene relación con otras variables.
MAR (Al azar): La falta depende de otra variable observada.
MNAR (No al azar): La falta depende del valor faltante en sí (ej. personas con altos ingresos no los reportan).

Métodos de Imputación:
Imputación Simple: Reemplazar por la media/mediana (datos numéricos) o moda (datos categóricos).
Imputación Avanzada: Uso de algoritmos de Machine Learning (como K-NN o regresión) para predecir el valor faltante.
Riesgos: Una mala imputación puede introducir sesgos y relaciones falsas en los datos. 

La imputación es fundamental para algoritmos de Machine Learning que no permiten valores nulos, permitiendo mantener el tamaño de la muestra.

```python
## Rellenar los valores faltantes en el df anterior, con el promedio
df_fill = df.fillna(
    {
        "Edad": df["Edad"].mean(),
        "Color" : "Desconocido"
    }
)

df_fill
```

```python
# Si en lugar de reemplazar los valores nulos queremos eliminarlos

df_sin_na = df.dropna()
df_sin_na
```
